```python
github_files = '../../../datasets/chipnemo_os/open_dataset/github_v.jsonl'
jsonObj = pd.read_json(path_or_buf=github_files, lines=True)
with open('github_repos_opensource.jsonl', 'w') as fout:
    for item in pd.unique(jsonObj['repo_name']).tolist():
        fout.write(json.dumps(item)+"\n")
jsonObj[0:100]
```
